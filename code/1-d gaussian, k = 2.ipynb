```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from sklearn.mixture import GaussianMixture as gmm
import math
```

```python
mu_1 = 2
variance_1 = 4
sigma_1 = math.sqrt(variance_1)

mu_2 = 10
variance_2 = 1
sigma_2 = math.sqrt(variance_2)

x_1 = np.linspace(mu_1 - 3*sigma_1, mu_1 + 3*sigma_1, 100)
x_2 = np.linspace(mu_2 - 3*sigma_2, mu_2 + 3*sigma_2, 100)
data_1 = norm.pdf(x_1, mu_1, sigma_1)
data_2 = norm.pdf(x_2, mu_2, sigma_2)

plt.plot(x_1, data_1, x_2, data_2)
plt.show()
```

```python
new_data = np.concatenate((data_1, data_2), axis=0)
np.random.shuffle(new_data)
new_data = new_data.reshape(-1, 1)
new_data.shape
```

```python
k = 2
model = gmm(n_components=k, max_iter=1000, covariance_type='full')
model.fit(new_data)

print(model.means_)
print(np.sqrt(model.covariances_))
print(np.mean(data_1), np.mean(data_2))
print(np.std(data_1), np.std(data_2))
```
